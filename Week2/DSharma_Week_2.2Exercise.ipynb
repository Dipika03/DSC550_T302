```python
# Importing the required libraries to plot graph

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Creating the data frame of only required columns STATE/UT and TOTAL

Road_Accident_Total_df = Road_Accidents_M_df[['STATE/UT','TOTAL']]
```

```python
# Using Group by to calculate the total road accidents by STATE/UT column

Road_Accident_Total_df = Road_Accident_Total_df.groupby(['STATE/UT']).sum()
```

### Bar graph

```python
# Let use the bar graph to understand which state has maximum road accidents.
Road_Accident_Total_df.plot(kind='bar')
# Add chart title
plt.title('Total Accidents by State', fontsize=14) 
 # Add x-axis label with font size of 14
plt.xlabel('State Name', fontsize=14)
# change rotation of x-axis labels
plt.xticks(rotation=90) 
# Add y-axis label with font size of 14
plt.ylabel('Total Accidents', fontsize=14) 
# Add grid lines 
plt.grid(True) 
plt.show()
```

#### From above bar plot we can clearly see that state Tamil Nadu in India has maximum road accidents during period 2001-2014.

```python
# Let create a data frame to see which year in tamil Nadu state has maximum number of road accidents.

Accidents_TN_df = Road_Accidents_M_df.loc[Road_Accidents_M_df['STATE/UT'] == "Tamil Nadu"] 
```

```python
# Creating the datafram for state Tamil nadu to show only required columns that is Year and Total.

Accidents_TN_df = Accidents_TN_df[['YEAR','TOTAL']]
```

```python
# Checking the data if it has relevant information

Accidents_TN_df
```

```python
# uaing the bar graph to see which year in state Tamil Nadu has maximum number of road accidents.
plt.bar(Accidents_TN_df['YEAR'], Accidents_TN_df['TOTAL'])
# Add chart title
plt.title('Total Accidents in Tamil Nadu by year ', fontsize=14) 
 # Add x-axis label with font size of 14
plt.xlabel('Year', fontsize=14)
# change rotation of x-axis labels
plt.xticks(rotation=90) 
# Add y-axis label with font size of 14
plt.ylabel('Total Accidents', fontsize=14) 
# Add grid lines 
#plt.grid(True) 
plt.show()
```

### As we can clearly see from above graph that year 2005 has maximum number of road accidents in Tamil nadu state

```python
# Lets use another dataset from kaggle website to understand at what time maximum number of road accidents happened in Tamil Nadu in 2005 year.

Accidents_TN2005_df = Road_Accidents_df.loc[Road_Accidents_df['STATE/UT'] == "Tamil Nadu"] 
```

```python
# checking the datafram data if is only showing Tamil nadu state data

Accidents_TN2005_df
```

```python
# using filter to show only 2005 year data for Tamil Nadu state.

Accidents_TN2005_df = Accidents_TN2005_df.loc[Accidents_TN2005_df['YEAR'] == 2005] 
```

```python
# To check if the created data frame showing correct data.

Accidents_TN2005_df
```

### Pie Chart

```python
# Lets use pie chart to see at what time maximum number of road accidents happened in tamil Nadu in 2005 year

Accidents_TN2005_pie_df=Accidents_TN2005_df[Accidents_TN2005_df.columns[3:10]].sum(axis=0)
Accidents_TN2005_pie_df.plot.pie(title='Time slot distribution of all accidents in Tamil Nadu in 2005',autopct='%.0f')
```

#### Above pie chart clearly shows that most of the road accidents in Tamil Nadu state in 2005 year happended during day time at 9-12 hrs and 15-18 hrs

```python
# Lets create new data frame using Year and total to see which Year has maximum number of road accidents.

Road_Accident_Year_df = Road_Accidents_M_df[['YEAR','TOTAL']]
```

```python
# using groupby function to calculate the total number of accident by Year.

Road_Accident_Year_df = Road_Accident_Year_df.groupby(['YEAR']).sum()
```

```python
# Lets see how data looks in newly created data frame

Road_Accident_Year_df
```

### Line Chart

```python
# Using Line chart to see what year has maximum number of road accidents.

plt.plot(Road_Accident_Year_df)
plt.title('Total Accidents by year ', fontsize=14) 
 # Add x-axis label with font size of 14
plt.xlabel('Year', fontsize=14)
# change rotation of x-axis labels
plt.xticks(rotation=90) 
# Add y-axis label with font size of 14
plt.ylabel('Total Accidents', fontsize=14) 
# Add grid lines 
#plt.grid(True) 
plt.show()
```

### The above Line graph clearly shows that year 2014 has maximum number of road accidents and the graph also show constant increase in road accidents every year.
